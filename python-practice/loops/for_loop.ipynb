```python
firstname = ['annas', 'ayan', 'qasim', 'ale', 'ibrahim']
lastname = ['pasha', 'hussain','hassan','irfan','javaid']
fullname = []
for first_name in firstname:
    for last_name in lastname:
        fullname.append(first_name + " " + last_name)
print(fullname)
```

```python
firstname = ['annas', 'ayan', 'qasim', 'ale', 'ibrahim']
lastname = ['pasha', 'hussain','hassan','irfan','javaid']
fullname = []

for first, last in zip(firstname, lastname):
    fullname.append(first + " " + last)

print(fullname)
```
